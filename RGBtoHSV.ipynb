```python
import time

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline
%load_ext autoreload
%autoreload 2
import numpy as np

import os, sys

from preprocessing import *
import image_preprocessing

from skimage import color

import keras

from skimage.color import rgb2hsv

import scipy

from keras.layers import Conv2D, Dense, Flatten, MaxPooling2D, LeakyReLU, Dropout
```

## Transform train images from rgb to hsv colorspace

```python
ROOT_DIR = 'training/'
IMAGE_DIR = ROOT_DIR + 'images/'
GT_DIR = ROOT_DIR + 'groundtruth/'
```

```python
files = os.listdir(IMAGE_DIR)

imgs = np.stack([load_image(IMAGE_DIR + file) for file in files]) # images (400, 400, 3)

hsv_imgs = [rgb2hsv(img) for img in imgs]
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(hsv_imgs[0])
plt.show()
```

```python
HSV_IMG_DIR = ROOT_DIR + 'hsv_images/'

for i, img in enumerate(hsv_imgs):
    scipy.misc.imsave(HSV_IMG_DIR + files[i], img)
```

## Transform test images from rgb to hsv colorspace
